A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. Please state the assumptions and tests that you carried out to check validity of the assumptions
'''

```python
import numpy as np
import pandas as pd
df=pd.read_csv("Cutlets.csv")
df
```

H0= There is no significant difference in diameters of cutlets from the two units;
H1= There is significant difference;
TWO-SAMPLE T-TEST;
Using scipy library;

```python
from scipy import stats
group1 = df['Unit A']
group2 = df['Unit B']
t_statistic, p_value = stats.ttest_ind(group1, group2)

print("t_statistic:",t_statistic.round(3))
print("P-value:",p_value.round(3))
aplha = 0.05

alpha = 0.05
if p_value < alpha:
  print("There is a significant difference in the diameter of the cutlet between two units, \n H1 is true")
else:
  print("There is no significant difference in the diameter of the cutlet between two units, \n H0 is true")
```

#Hypothesis testing Exercise-2
'''
A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.

Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.
'''

```python
from google.colab import files
uploaded = files.upload()
```

```python
import numpy as np
import pandas as pd
df=pd.read_csv("LabTAT.csv")
df
```

When we have more than two groups (in this case, four laboratories)
and want to compare their means
to see if there are any significant differences,
we have to use ANOVA-test

```python
lab1=df["Laboratory 1"]
lab2=df["Laboratory 2"]
lab3=df["Laboratory 3"]
lab4=df["Laboratory 4"]
from scipy import stats
f_statistic, p_value = stats.f_oneway(lab1,lab2,lab3,lab4)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in average TAT among the laboratories.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in average TAT among the laboratories.")
```

# **Hypothesis testing Exercise-3:**
Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions.

```python
from google.colab import files
uploaded = files.upload()
```

```python
import numpy as np
import pandas as pd
df=pd.read_csv("BuyerRatio.csv")
df
```

Chi-squared test of independence to determine if male-female buyer ratios differ across regions, as it assesses categorical data associations.

```python
from scipy.stats import chi2_contingency

observed_values = df.iloc[:, 1:].values.astype(int)
print(observed_values)


chi2, p, _, _ = chi2_contingency(observed_values)

alpha = 0.05

if p < alpha:
    print("Reject the null hypothesis: Male-female buyer ratios are not similar across regions.")
else:
    print("Fail to reject the null hypothesis: Male-female buyer ratios are similar across regions.")


```
